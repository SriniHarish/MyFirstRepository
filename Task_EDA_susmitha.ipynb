**--------------------------------------------------------1ST METHOD---------------------------------------------------------------**

**find the missing values**

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()

# look at the # of missing points in the all columns
missing_values_count[0:44]
```

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()
```

**percent of data that is missing**

```python
# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
#almost half of the data are missing
```

```python
# remove all the rows that contain a missing value
df.dropna()
```

```python
#if we try to remove the row with missing value then all the records get 
#dropped which indicates every row have some missing data
```

```python
# remove all columns with at least one missing value

columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head()
```

#**---------------------------------------------------2ND METHOD------------------------------------------------------------**

```python
df.head()
```

```python
#drop parenet id column
after_drop=df.drop(['parent_id'], axis=1)
after_drop
```

```python
#column reduced to 43
```

**data imputation**

```python
#here relationships value is filled with mean value using 
after_drop['relationships'].fillna(after_drop['relationships'].mean(), inplace=True)

after_drop['lat'].fillna(after_drop['lat'].mean(), inplace=True)

after_drop['lng'].fillna(after_drop['lng'].mean(), inplace=True)


after_drop
```

```python
categorical_col = after_drop.select_dtypes(include = ["object"]).columns.values
after_drop[categorical_col] = after_drop[categorical_col].fillna("Unknown")
len(categorical_col)
```

```python
# Numeric datatype Variables set empty value to 0

numeric_col = after_drop.select_dtypes(exclude = ["object"]).columns.values
after_drop[numeric_col] = after_drop[numeric_col].fillna(0)
after_drop.select_dtypes(exclude = ["object"]).isnull().sum()
```
